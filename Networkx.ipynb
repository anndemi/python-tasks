# [NetworkX](https://www.google.com/url?q=https%3A%2F%2Fnetworkx.org%2F) - библиотека для работы с графами и сетями.

В рассмотренном далее примере с помощью генератора  построен полный граф $K_5$ (граф является ***полным***, если в нём все вершины соединены друг с другом рёбрами). Полный список генераторов различных графов в NetworkX можно найти [здесь](https://networkx.org/documentation/stable/reference/generators.html), однако создавать новые вершины и рёбра можно и вручную.

В рассматриваемом примере граф $G$ имеет 5 вершин с номерами от 0 до 4, а граф $H$ получается из него добавлением новой вершины под номером 5 и ребра, соединяющего эту вершину с вершиной 3. Расположение вершин выбирается произвольно через ```random_layout``` (про различные укладки можно прочитать в [документации](https://networkx.org/documentation/stable/reference/drawing.html#module-networkx.drawing.layout)).

С заданием таких параметров, как цвета вершин и рёбер, размер вершин и т.д. можно ознакомиться [здесь](https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw_networkx.html#draw-networkx).

````python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.complete_graph(5)
pos = nx.random_layout(G) #
options = {"edge_color": "blue", "node_size": 200, "node_color": "red"}
plt.figure(3, figsize=(10, 10))
plt.subplot(221)
nx.draw_networkx(G, pos, **options, with_labels = True) #
plt.subplot(222)
H = G
H.add_node(5)
H.add_edge(3,5)
pos1=nx.random_layout(H)
nx.draw_networkx(H, pos1, **options, with_labels = True)
````

**Определение:** ***Степень вершины*** неориентированного графа – это количество инцидентных ей рёбер.

**Определение:** Граф называется ***помеченным*** (перенумерованным), если его вершины отличаются какими-либо пометками.

Пусть $G$ – помеченный граф с $n$ вершинами.

**Определение:** ***Матрица смежности*** $A=(a_{i,j})$ графа $G$ – бинарная (состоящая из 0 и 1) матрица размера $n \times n$ следующего вида:

$a_{i,j}=1$, если вершины $i$ и $j$ – смежные (т.е. соединены ребром), и $a_{i,j}=0$ в противном случае.

Матрица смежности является симметрической ($a_{i,j}=a_{j,i}$ $\forall i, j$). Число единиц в строке совпадает со степенью вершины, соответствующей данной строке.

Построить матрицу смежности можно с помощью метода `nx.adjacency_matrix` (полученный объект будет иметь тип ```numpy.ndarray```).
