# Визуализация булева куба с использованием библиотеки NetworkX

***Булев куб***  размерности $n$ - множество всех двоичных (состоящих из 0 и 1) наборов длины $n$.

В библиотеке NetworkX есть генератор `nx.cubical_graph()`, позволяющий нарисовать трёхмерный куб.

```python
import networkx as nx
import matplotlib.pyplot as plt

H = nx.cubical_graph()
nx.draw_networkx(H)
```

Тем не менее, в булевом кубе вершинам соответствуют двоичные наборы, и для задания соответствующих меток удобнее прописать их вручную. В следующем примере реализовано размещение на плоскости вершин, соответствующих двоичным наборам $000$, $001$, $010$ и $100$, а также построение рёбер, соединяющих вершину с набором $000$ с соседними для неё в булевом кубе.

```python
side = 4
step = 2
options = {"node_size": 200}

cube = nx.Graph()
cube.add_node("000", pos_c=(0, 0))
cube.add_node("001", pos_c=(side, 0))
cube.add_node("010", pos_c=(0, side))
cube.add_node("100", pos_c=(step, 0+step))

all_edges = [("000", "001"), ("000", "010"), ("000", "100")]

cube.add_edges_from(all_edges)
pos=nx.get_node_attributes(cube,'pos_c')
plt.figure(3, figsize=(4, 4))
nx.draw(cube, pos, **options, node_color="white")
nx.draw_networkx_labels(cube, pos, font_size=9, font_color="black")
```
